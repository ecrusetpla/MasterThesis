```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
```

```python
p = "../data/population/Population and migration by postal code (1998-2023).csv"
df = pd.read_csv(p, sep=";", encoding="utf-8-sig")
df.head()
```

```python
df = df.replace(".", np.nan)
for c in df.columns:
    df[c] = pd.to_numeric(df[c], errors="ignore")
df.dtypes.head(15)
```

```python
df.shape, df.isna().mean().sort_values(ascending=False).head(15)
```
